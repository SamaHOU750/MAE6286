```python
import numpy
from matplotlib import pyplot

%matplotlib inline
```

```python
def linear_convection(nx, L=2.0,c=1.0,dt=0.025, nt=20):
    dx=L/(nx-1)
    x=numpy.linspace(0.0, L,num=nx)
    u0=numpy.ones(nx)
    mask=numpy.where((x>=0.5)&(x<=1.0))
    u0[mask]=2.0
    u=u0.copy()
    for n in range(nt):
        un= u.copy()
        u[1:]=un[1:] - c * dt/ dx* (un[1:]-un[:-1])
    pyplot.figure() 
    pyplot.grid()
    pyplot.plot(x, u0, label='initial')
    pyplot.plot(x, u, label='Current')
    pyplot.legend()
    pyplot.xlim(0.0,L)
    pyplot.xlim(0.0, 2.5)
        
```

```python
linear_convection(41)
```

```python
def linear_convection_cf(nx, L=2.0,c=1.0,sigma=0.5, nt=20):
    dx=L/(nx-1)
    x=numpy.linspace(0.0, L,num=nx)
    dt=sigma*dx/c
    u0=numpy.ones(nx)
    mask=numpy.where((x>=0.5)&(x<=1.0))
    u0[mask]=2.0
    u=u0.copy()
    for n in range(nt):
        un= u.copy()
        u[1:]=un[1:] - c * dt/ dx* (un[1:]-un[:-1])
    pyplot.figure() 
    pyplot.grid()
    pyplot.plot(x, u0, label='initial')
    pyplot.plot(x, u, label='Current')
    pyplot.legend()
    pyplot.xlim(0.0,L)
    pyplot.xlim(0.0, 2.5)
```

```python
linear_convection_cf(85)
```

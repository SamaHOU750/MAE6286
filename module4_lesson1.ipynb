```python
import numpy
from matplotlib import pyplot

%matplotlib inline
```

```python
nx = 62
sigma = 0.5
L = 1
dx = L/(nx-1)
alpha = 1.22*10**-3
dt = sigma *dx**2/alpha
nt = 100
nt_2 = 1000
x = numpy.linspace(0,L,num = nx)

T0 = numpy.zeros(nx)
T0[0] = 100
```

```python
def FTCS(T0,nt,dt,dx,alpha):
    sigma = alpha * dt/(dx**2)
    T = T0.copy()
    for n in range(nt):
        T[1:-1] = T[1:-1] + sigma * (T[:-2] - 2 * T[1:-1] + T[2:])
    return T
```

```python
T = FTCS(T0, nt, dt, dx, alpha)
```

```python
pyplot.figure(figsize=(10.0, 10.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [C]')
pyplot.grid()
pyplot.plot(x, T, color='C0', linestyle='-', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100.0);
```

```python
def FTCS2(T0,nt_2,dt,dx,alpha):
    sigma = alpha * dt/(dx**2)
    T2 = T0.copy()
    for n in range(nt):
        T2[1:-1] = T2[1:-1] + sigma * (T2[:-2] - 2 * T2[1:-1] + T[2:])
    return T2
```

```python
T2 = FTCS2(T0, nt_2, dt, dx, alpha)
```

```python
pyplot.figure(figsize=(10.0, 10.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [C]')
pyplot.grid()
pyplot.plot(x, T2, color='C1', linestyle='-', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100.0);
```
